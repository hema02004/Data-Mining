```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris_df = sns.load_dataset('iris')  # Load the Iris dataset using Seaborn
sns.FacetGrid(iris_df, hue="species", height=2).map(sns.distplot, "petal_length").add_legend()
sns.FacetGrid(iris_df, hue="species", height=2).map(sns.distplot, "petal_width").add_legend()
sns.FacetGrid(iris_df, hue="species", height=2).map(sns.distplot, "sepal_length").add_legend()
sns.FacetGrid(iris_df, hue="species", height=2).map(sns.distplot, "sepal_width").add_legend()
plt.show()
```

```python
iris=load_iris()
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)


# Initialize the Decision Tree Classifier
clf = DecisionTreeClassifier()

# Train the classifier on the training data
clf.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = clf.predict(X_test)

# Evaluate the model
accuracy = clf.score(X_test, y_test)
print("Accuracy:", accuracy)

# Plot the decision tree
plt.figure(figsize=(12, 8))
plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
```
